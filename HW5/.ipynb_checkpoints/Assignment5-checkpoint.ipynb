```python
import geopandas as gp
import pandas as pd
import os
import json
import pylab as pl
s = json.load( open(os.getenv('PUI2015')+'pui_plot.json') )
pl.rcParams.update(s)
%pylab inline
```

```python
# loading data frame
nrg = gp.GeoDataFrame.from_csv("../../Week5/Energy_and_Water_Data_Disclosure_for_Local_Law_84__2013_.csv")
print nrg.columns
```

```python
#second data frame:
bsize = gp.GeoDataFrame.from_file("../../Week5/Manhattan/MNMapPLUTO.shp")
print bsize.columns
```

```python
#plotting the scatter matrix:
from pandas.tools.plotting import scatter_matrix
scatter_matrix (nrg, s=300, figsize=(16, 16), diagonal='kde')
```

```python
#Making a function that figures out whether an attribute can be converted to float or not
def canconvert(mydata):
    try :
        float(mydata)
        return True
    except ValueError:
        return False
```

```python
import numpy as np
#create the indeces array here
findx = np.vectorize(canconvert)
indx = findx(nrg['Site EUI(kBtu/ft2)'])
nrg['Site EUI(kBtu/ft2)'][~indx]=float('NaN')
```

```python
indx = findx(nrg['Reported Property Floor Area (Building(s)) (ft²)'])
nrg['Reported Property Floor Area (Building(s)) (ft²)'][~indx]=float('NaN')
```

```python
#Checking that it worked:
print nrg['Site EUI(kBtu/ft2)'].astype(float)
print nrg['Reported Property Floor Area (Building(s)) (ft²)'].astype(float)
print bsize.BBL.astype(float)
```

```python
#adding more columns, coz why waste all the extra RAM?
nrg['Energy'] = nrg['Site EUI(kBtu/ft2)'].astype(float)
nrg['Floor_Area'] = nrg['Reported Property Floor Area (Building(s)) (ft²)'].astype(float)
```

```python
#Merging
nrg['BBL'] = nrg['NYC Borough, Block, and Lot (BBL)']

#merge the table file with the shape file
BBLmerge = pd.merge(nrg, bsize, on='BBL')
BBLmerge['Building_Energy_Consumption'] = BBLmerge['Energy']*BBLmerge['Floor_Area']
```

```python
BBLmerge.plot(kind='scatter', x='UnitsTotal', y='Building_Energy_Consumption', color='b', alpha=0.8)
#plt.scatter(x=np.log10(BBLmerge['UnitsTotal']),y=np.log10(BBLmerge['Building Energy Consumption']),linestyle='o', color='b', alpha=0.8)
pl.xlim([1, 1000])
pl.ylim([1000, 1e10])
pl.xlabel('Total Units')
pl.ylabel('Total Energy (kBtu)')
pl.grid()
pl.legend()
```

```python
BBLmerge.plot(kind='scatter', x='UnitsTotal', y='Building_Energy_Consumption', color='b', alpha=0.8, loglog=True)
#plt.scatter(x=np.log10(BBLmerge['UnitsTotal']),y=np.log10(BBLmerge['Building Energy Consumption']),linestyle='o', color='b', alpha=0.8)
pl.xlim([1, 1000])
pl.ylim([1000, 1e10])
pl.xlabel('Total Units')
pl.ylabel('Total Energy (kBtu)')
pl.grid()
pl.legend()
```

```python
# Cutting the dataset

BBLCut = BBLmerge[(BBLmerge['Building_Energy_Consumption']>=1000)&(BBLmerge['Building_Energy_Consumption']<1e+10) & (BBLmerge['UnitsTotal']>=1)&(BBLmerge['UnitsTotal']<1000)]

BBLCut
```

```python
BBLCut.plot(kind='scatter', x='UnitsTotal', y='Building_Energy_Consumption', color='b', alpha=0.8, loglog=True)
pl.xlim([1, 1000])
pl.ylim([1000, 1e10])
pl.xlabel('Total Units')
pl.ylabel('Total Energy (kBtu)')
pl.grid()
pl.legend()
```

```python
import statsmodels.api as sma

y = (BBLCut['Building_Energy_Consumption'])
x = (BBLCut['UnitsTotal'])
#Fitting a line through the data
x1 = sma.add_constant(np.log10(x))
y1 = np.log10(y) 

model1 = sma.OLS(y1, x1, missing='drop').fit()
print model1.summary()
```

```python
figure1 = figure(figsize = (10,10))
xPlot1 = np.arange(0, 4, 1)
yPlot1 = (model1.params)[1]*xPlot1 + (model1.params)[0]
ax = figure1.add_subplot(111)
ax.plot(x1, y1, 'ko')
ax.plot(xPlot1, yPlot1, 'red')
plt.xlabel('Total Units (Log value)')
plt.ylabel('Total Energy Consumption (Log value)')
```

```python
#Swapping the axes to see if fit is improved
x2 = sma.add_constant(np.log10(y))
y2 = np.log10(x)

model2 = sma.OLS(y2, x2, mising='drop').fit()
print model2.summary()

x2 = (np.log10(y))
y2 = sma.add_constant(np.log10(x))
```

```python
figure2 = figure(figsize = (10,10))
xPlot2 = np.arange(1, 10, 1)
yPlot2 = (model2.params)[1]*xPlot2 + (model2.params)[0]
ax = figure2.add_subplot(111)
ax.plot(x2, y2, 'yo')
ax.plot(xPlot2, yPlot2, 'purple')
plt.xlabel('Total Energy Consumption (Log value)')
plt.ylabel('Total Units (Log value)')
```
