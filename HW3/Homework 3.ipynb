```python
from __future__ import print_function
__author__= 'amp914'
import os
import numpy as np
import pylab as pl
%pylab inline

from scipy.optimize import curve_fit, minimize
```

```python
distributions = ['pois', 'gaus', 'chisq', 'gamma', 'binomial', 'lognormal']
mymean=100
df = 100
mysize=1000/(np.array(range(1,100)+[10]))
```

### Poisson

```python
i=0
distributions[i] = {}
for n in mysize:
    distributions[i][n] = np.random.poisson(mymean, size=n)
means = {}
means[i] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in distributions[i].iterkeys():
    means[i][nn] = distributions[i][nn].mean()
    #and plot it
    axchisq_mu_n.plot(nn, means[i][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')
    
allmeans=[]
for n in means[i]:
    allmeans.append(means[i][n])
pl.figure(figsize=(10,10))
pl.hist(allmeans,bins=30)
pl.xlabel('sample mean')
pl.ylabel('N')
```
